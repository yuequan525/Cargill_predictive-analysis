**Hypothesis test 3** 
Transactions impact

<li><font color=blue> In this step, the transactions impact is checked by checking the weighted average unit price along the date for each item_class_l2 based group. No time trend is identified through the chart. However,by digging into the sample data for the last recent transcations, the similar transactions are found for each car. Therefore, the rolling average price taken for the last 4 transcations is calculated. If there is no more than 4 transcations for each car, the last price is taken into account as rolling average price.</li></font>

```python
## sort the data by date
df_v1 = df_v1.sort_values(by="date_of_sale")
date_tmp = pd.DataFrame(df_v1['date_of_sale']).drop_duplicates()
date_tmp["seq_date"] = np.arange(1, len(date_tmp) + 1)

df_car_by_date = df_v1.merge(date_tmp, on='date_of_sale', how='left')

## check the weighted average unit price against time

date_check = data_analysis(df_car_by_date)

wtavg_date = date_check.weighted_mean("no_of_units")

df_car_transactions = (df_car_by_date.groupby(['item_class_l2','date_of_sale','seq_date'])
                       .agg({'unit_price':[wtavg_date]}))

df_car_transactions = df_car_transactions.reset_index().loc[:,['item_class_l2','seq_date','unit_price']]

## plot by group
bp = df_car_transactions.groupby('item_class_l2').plot(x="seq_date", y="unit_price")
```

```python
### take the moving average
df_rolling = df_car_by_date.loc[:,['item_class_l2','car_nr','unit_price','date_of_sale','seq_date']]

df_rolling_avg= (df_rolling.sort_values(['car_nr','seq_date'],ascending=True)
                 .groupby('car_nr')['unit_price'].rolling(4).mean()
                 .reset_index()
                 .rename(columns={'car_nr':'car_nr','level_1':'seq_date','unit_price':'rolling_price'}))

df_rolling_price = df_car_by_date.merge(df_rolling_avg,on=["car_nr",'seq_date'],how="left").loc[:,['car_nr','unit_price','seq_date',
                                           'rolling_price']].drop_duplicates()

df_rolling_price.rolling_price.fillna(df_rolling_price.unit_price, inplace=True)

df_rolling_price = df_rolling_price.sort_values("seq_date").drop_duplicates('car_nr',keep='last')

df_v2 = df_car_by_date.merge(df_rolling_price.loc[:,['car_nr','rolling_price']],on=["car_nr"],how="left")
```

## Feature Engineering

**Step 1: Encode the categorical variables into numerical variables**
<li><font color = blue> The target-based encoding is used in this step. 
<li><font color = blue> Target-based encoding is numerization of a categorical variables via the target variable. Main purpose is to deal
    with high cardinality categorical features without exploding dimensionality. This replaces the categorical variable
    with just one new numerical variable. Each category or level of the categorical variable is represented by a
    summary statistic of the target for that level. Addition noise is included for overcoming the overfitting. 

```python

df_recoded = df_car
encode = data_analysis(df_recoded)

df_recoded['cust_segment_name'] = encode.target_encoder('cust_segment_name','cust_segment_name','unit_price',None,'mean')
df_recoded['item_class_l2'] = encode.target_encoder('item_class_l2','item_class_l2','unit_price',None,'mean')
df_recoded['item_class_l3'] = encode.target_encoder('item_class_l3',['item_class_l2','item_class_l3'],'unit_price',None,'mean')
df_recoded['expensive_part_cluster'] = encode.target_encoder('expensive_part_cluster','expensive_part_cluster','unit_price',None,'mean')
df_recoded['destination'] = encode.target_encoder('destination','destination','unit_price',None,'mean')
```

```python
target_col= 'unit_price'
X = df_model.loc[:,df_model.columns != target_col]
y = df_model.loc[:,target_col]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=40)
X_with_constant = sm.add_constant(X_train)
lr_model = sm.OLS(y_train,X_with_constant)
result = lr_model.fit()
print(result.summary())
```

```python
X_test = sm.add_constant(X_test)
y_pred = result.predict(X_test)
any(n < 0 for n in y_pred)
residual = y_test - y_pred
```

```python
## check whether residuals are normal distributed
sns.distplot(residual)
fig,ax = plt.subplots(figsize=(6,2.5))
_,(__,__,r) = stats.probplot(residual,plot=ax,fit=True)

ax.scatter(y_pred,residual)

acf = smt.graphics.plot_acf(residual,lags=40,alpha=0.05)
acf.show()
```

<font color = blue> Firstly, we see that the model relationship is not exactly linear and there is non-constant variance in residuals after checking the residuals of the model. Secondly, there are negative values in the prediction. Taking the log of Y might solve the issue. However, we need to compare the predicted price with actual price and the predicted revenue and actual revenue. Taking the log transformation and transfor it back would cause the bias on the comparision. Therefore, nonparametric models are prefered for next step.
