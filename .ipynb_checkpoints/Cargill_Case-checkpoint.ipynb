
<li>Calculate the statistics for weighted average unit price based on the distribution of item_class_l2 and l3 </li>


```python
## get the statistic for class_l2 & l3

wtavg = lambda x: np.average(x, weights = df.loc[x.index, "no_of_units"],axis=0)
wtavg.__name__ = "weighted_avg"
f_stat = {'unit_price':['min','max',wtavg]}

## -------stat for l2 class vs. unit price ----##
class_l2_stat = df.groupby('item_class_l2').agg(f_stat)
class_l2_stat.columns = ['min_unit_price','max_unit_price','avg_unit_price']
class_l2_stat = class_l2_stat.reset_index().sort_values(by=['avg_unit_price'], ascending=True)

## -------stat for l3 class vs. unit price ----##

class_l3_stat = df.groupby(['item_class_l2','item_class_l3']).agg(f_stat)
class_l3_stat.columns = ['min_unit_price','max_unit_price','avg_unit_price']
class_l3_stat = class_l3_stat.reset_index().sort_values(by=['item_class_l2','avg_unit_price'], ascending=True)

print(class_l2_stat)
#print(class_l3_stat)
```

**Hypothesis test 3** 
Transactions impact

<li> In this step, the transactions impact is checked by checking the weighted average unit price along the date for each item_class_l2 based group. No time trend is identified through the chart. However,by digging into the sample data for the last recent transcations, the similar transactions are found for each car. Therefore, the rolling average price taken for the last 4 transcations is calculated. If there is no more than 4 transcations for each car, the last price is taken into account as rolling average price. The new constructed variable, rolling average price, is kept for further research.</li>

```python
## sort the data by date
df_v1 = df_v1.sort_values(by="date_of_sale")
date_tmp = pd.DataFrame(df_v1['date_of_sale']).drop_duplicates()
date_tmp["seq_date"] = np.arange(1, len(date_tmp) + 1)

df_car_by_date = df_v1.merge(date_tmp, on='date_of_sale', how='left')

## check the weighted average unit price against time

date_check = data_analysis(df_car_by_date)

wtavg_date = date_check.weighted_mean("no_of_units")

df_car_transactions = (df_car_by_date.groupby(['item_class_l2','date_of_sale','seq_date'])
                       .agg({'unit_price':[wtavg_date]}))

df_car_transactions = df_car_transactions.reset_index().loc[:,['item_class_l2','seq_date','unit_price']]

## plot by group
bp = df_car_transactions.groupby('item_class_l2').plot(x="seq_date", y="unit_price")
```

```python
### take the moving average
df_rolling = df_car_by_date.loc[:,['item_class_l2','car_nr','unit_price','date_of_sale','seq_date']]

df_rolling_avg= (df_rolling.sort_values(['car_nr','seq_date'],ascending=True)
                 .groupby('car_nr')['unit_price'].rolling(4).mean()
                 .reset_index()
                 .rename(columns={'car_nr':'car_nr','level_1':'seq_date','unit_price':'rolling_price'}))

df_rolling_price = df_car_by_date.merge(df_rolling_avg,on=["car_nr",'seq_date'],how="left").loc[:,['car_nr','unit_price','seq_date',
                                           'rolling_price']].drop_duplicates()

df_rolling_price.rolling_price.fillna(df_rolling_price.unit_price, inplace=True)

df_rolling_price = df_rolling_price.sort_values("seq_date").drop_duplicates('car_nr',keep='last')

df_v2 = df_car_by_date.merge(df_rolling_price.loc[:,['car_nr','rolling_price']],on=["car_nr"],how="left")
```

## Feature Engineering

**Step 1: Encode the categorical variables into numerical variables**
<li> The target-based encoding is used in this step. 
<li> Target-based encoding is numerization of a categorical variables via the target variable. Main purpose is to deal
    with high cardinality categorical features without exploding dimensionality. This replaces the categorical variable
    with just one new numerical variable. Each category or level of the categorical variable is represented by a
    summary statistic of the target for that level. Addition noise generated from normal distribution is included for overcoming the overfitting. 

```python

df_recoded = df_car
encode = data_analysis(df_recoded)

df_recoded['cust_segment_name'] = encode.target_encoder('cust_segment_name','cust_segment_name','unit_price',None,'mean')
df_recoded['item_class_l2'] = encode.target_encoder('item_class_l2','item_class_l2','unit_price',None,'mean')
df_recoded['item_class_l3'] = encode.target_encoder('item_class_l3',['item_class_l2','item_class_l3'],'unit_price',None,'mean')
df_recoded['expensive_part_cluster'] = encode.target_encoder('expensive_part_cluster','expensive_part_cluster','unit_price',None,'mean')
df_recoded['destination'] = encode.target_encoder('destination','destination','unit_price',None,'mean')
```

**Step 2: transform categorical & numerical variables**

## Model
1. Check whether linearity exsits and linear regression is applicable 
2. Train the model and tuning the parameters via nested cross validation
3. Select the final model for prediction

```python
df_model = df_recoded.drop(columns=['no_of_units',"item_class_l3","delivery_duration"])
```

```python
df_model
```

**Linear Regression**

```python
target_col= 'unit_price'
X = df_model.loc[:,df_model.columns != target_col]
y = df_model.loc[:,target_col]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=40)
X_with_constant = sm.add_constant(X_train)
lr_model = sm.OLS(y_train,X_with_constant)
result = lr_model.fit()
print(result.summary())
```

```python
X_test = sm.add_constant(X_test)
y_pred = result.predict(X_test)
any(n < 0 for n in y_pred)
residual = y_test - y_pred
```

**check whether residuals are normal distributed**

```python
sns.distplot(residual)
fig,ax = plt.subplots(figsize=(6,2.5))
_,(__,__,r) = stats.probplot(residual,plot=ax,fit=True)

ax.scatter(y_pred,residual)

acf = smt.graphics.plot_acf(residual,lags=40,alpha=0.05)
acf.show()
```

Firstly, we see that the model relationship is not exactly linear and there is non-constant variance in residuals after checking the residuals of the model. Secondly, there are negative values in the prediction. Taking the log of Y might solve the issue. However, we need to compare the predicted price with actual price and the predicted revenue and actual revenue. Taking the log transformation and transfor it back would cause the bias on the comparision. Therefore, nonparametric models are prefered for next step. 
   

```python
%run Model.py
test = Model(df_model)

dataX = test.dataX(["cust_segment_name","item_class_l2",
                    "expensive_part_cluster","volume",
                    "rolling_price","destination"])
dataY = test.dataX(["unit_price"])
```

**KNN**

```python
#### KNN ####
KNN_trained = test.train_model_grid(neighbors.KNeighborsRegressor(),
                                    [{'n_neighbors': [2,3,4,5,6], 'weights': ['uniform','distance']}],
                                    dataX,dataY)
```

**Random Forest**

```python
##### Random Forest #####
rf = RandomForestRegressor(random_state = 42)

# Create the random grid
# Number of trees in random forest
n_estimators = [10,20,40]
# Number of features to consider at every split
max_features = ['sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(1, 10, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2,3,4,5,6]
# Minimum number of samples required at each leaf node
min_samples_leaf = [int(x) for x in np.linspace(100, 400, num = 4)]
 
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

pprint(random_grid)

###NOTE: as the computeration pwoer of my own laptop, I cann't take the full combination of parameters I planned to
###      tune the model, therefore I select the small combination for cross validation.(Mainly regarding max_deapth) Further tuning parameters 
###      could improve the model performance. 

max_depth = [4]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

RF_trained = test.train_model_grid(rf,random_grid,dataX,dataY)
```

**Gradient Boosting**

```python
#### Gradient Boosting ####
GB = GradientBoostingRegressor(random_state=42)
pprint(GB.get_params())

gb = GradientBoostingRegressor(learning_rate=0.1,
                                max_features='sqrt', 
                                subsample=0.8, random_state=42)

gb_grid = {'max_depth':[4], 
           'min_samples_split':min_samples_split,
           'n_estimators': n_estimators,
           'min_samples_leaf': min_samples_leaf}

###NOTE: as the computeration pwoer of my own laptop, I cann't take the full combination of parameters I planned to
###      tune the model, therefore I select the small combination for cross validation. Further tuning parameters 
###      could improve the model performance. 

GB_trained = test.train_model_grid(gb,gb_grid,dataX,dataY)
```

After comparison, Gradient Boosting and KNN model are selected

The accuracy of KNN model is higher. However, considering that KNN just searches for nearest neighbors in the data. It might be influenced by extreme values. Gradient Boosting is preferred for final prediction step.
